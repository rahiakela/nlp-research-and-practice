##Word vector representation

image.png

Let's build a vector for input text(doc1 and doc2).

```python
doc1 = """meeting ... management ... meeting ... management ... meeting ... management ... meeting ... meeting"""

doc1_vector = [0, 0]

for word in doc1.split(" "):
  if word == "management":
    doc1_vector[0] += 1
  if word == "meeting":
    doc1_vector[1] += 1

print(doc1_vector)
```

```python
doc2 = """management ... meeting ... management ... management ... management"""

doc2_vector = [0, 0]

for word in doc2.split(" "):
  if word == "management":
    doc2_vector[0] += 1
  if word == "meeting":
    doc2_vector[1] += 1

print(doc2_vector)
```

image.png

Let's build a vector for query text.

```python
query = "management meeting"

query_vector = [0, 0]

for word in query.split(" "):
  if word == "management":
    query_vector[0] += 1
  if word == "meeting":
    query_vector[1] += 1

print(query_vector)
```

image.png

Can we measure the distance between each pair in precise terms?

The measurement of distance in geometrical space originates with the good old
Pythagorean theorem.

This calculation is called Euclidean distance, and the geometrical interpretation is generally referred to as Euclidean space.

Let's calculate Euclidean distance for `doc1` and `doc2` with respect to query vector.

```python
import math

def square_length(doc, query):
  square_length = 0
  for index in range(0, len(query)):
    square_length += math.pow((doc[index] - query[index]), 2)
  return round(math.sqrt(square_length), 3)
```

```python
print(f"Doc1 Euclidean distance: {square_length(doc1_vector, query_vector)}")
```

```python
print(f"Doc1 Euclidean distance: {square_length(doc2_vector, query_vector)}")
```

Note that if we typed in management and meeting multiple times in our query,
the content and information need would not change, but the vector itself would.

In particular,
the length of the vector will be different, but the angle between the first version
of the vector and the second one won’t change.

image.png

```python
query = "management meeting management meeting"

query_vector = [0, 0]

for word in query.split(" "):
  if word == "management":
    query_vector[0] += 1
  if word == "meeting":
    query_vector[1] += 1

print(query_vector)
```

```python
print(f"Doc1 Euclidean distance: {square_length(doc1_vector, query_vector)}")
```

```python
print(f"Doc1 Euclidean distance: {square_length(doc2_vector, query_vector)}")
```

Therefore, it is much more informative to measure the angle between the length normalized vectors rather than the absolute distance, which can be dependent on the length of the documents.
